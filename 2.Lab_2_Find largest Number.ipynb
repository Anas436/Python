You are given three numbers from keyboard. Find the largest one?

```python
x=int(input("Enter a first number : "))
y=int(input("Enter a Second Number : "))
z=int(input("Enter a Third Number : "))

largest =0

if x > y & x > z :
  largest=x

elif y > z:
  largest=y
else:
  largest =z

print(largest, "it's a largest number !")
```
